```python
def convert_to_float(value):
    if isinstance(value, str): 
        return float(value.replace(',', '.')) 
    return float(value)

columns_to_convert = ['Surface Carrez du 1er lot', 'Surface Carrez du 2eme lot', 'Surface Carrez du 3eme lot', 'Surface Carrez du 4eme lot', 'Surface Carrez du 5eme lot']
for column in columns_to_convert:
    df[column] = df[column].apply(convert_to_float)
    
```

```python
# 4/ Proportion d’appartements vendus par nombre de pièces
sql_query = '''
SELECT biens.nb_pieces_principales, count(biens.nb_pieces_principales) / (
    SELECT count(*)
    FROM ventes
    JOIN biens ON biens.id_biens = ventes.ventes_id
    where type_local='Appartement' 
) * 100 as proportion
FROM ventes
JOIN biens ON biens.id_biens = ventes.ventes_id
where type_local = 'appartement'
group by biens.nb_pieces_principales
Limit 50;
'''
res = pd.read_sql(sql_query, engine)
res
```
